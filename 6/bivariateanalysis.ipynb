### 1.Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(data = tips, x='total_bill', y='tip')
plt.title('Scatter plot of Total Bill vs Tip')
plt.show()
```

###### this is linear , the amount of the bill mostly decide how much the tip is going to be

```python
# can even add a regression line or color-code by a third variable hue 
sns.scatterplot(data=tips , x= 'total_bill' , y= 'tip' , hue = 'sex')
```

#### multivariate analysis

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', style='smoker', markers={'Yes': '^', 'No': 'o'} ,hue = 'sex' ,size = 'size')
```

### 2.Bar Plot(Numerical - Categorical)

```python
# using tiatanic dataset
titanic.sample(5)
```

```python
# generally in barplot X->categorical data , y-> numerical data
sns.barplot(data=titanic , x='Pclass' , y='Age' , hue='Sex' )
plt.show()
```

```python
sns.barplot(data=titanic, x = 'Survived' , y ='Fare' ,hue = 'Sex')
plt.plot()
```

###### can clearly see that the higher the fare was higher chance of survival 

### 3.Box Plot (Numerical - Categorical)

```python
sns.boxplot(data = titanic , x= 'Sex' , y= 'Age', hue = 'Survived')
```

### 4.Distplot(Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'] )
sns.distplot(titanic[titanic['Survived'] == 1]['Age'] )
```
