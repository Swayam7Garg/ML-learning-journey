```python
import pandas as pd
import numpy as np
```

Series  : is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). 

```python
a = {1 : 'a' , 2: 'x' , 3 : 'g'}
pd.Series(a , index =[1 , 2,5 ,3])
```

Dataframe : 2 d labelled data structure with columns of pottentially different datatype

```python
# from series
d = { "one" : pd.Series([1,2,3] , index =["a" , "b" , "c"]),
     "two" : pd.Series([4,5,6] , index =["x" , "y" , "z"])}
df = pd.DataFrame(d)
```
