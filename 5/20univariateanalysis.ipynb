# 1.categorical data

### Countplot

```python
sns.countplot(x = df['Survived'])
df['Survived'].value_counts()
```

```python
sns.countplot(x=df['Sex'])
```

```python
sns.countplot(x=df['Embarked'])
```

### b.Piechart

```python
# we need to draw a piechart to get the information in percentage 
# we can draw a piechart through matplotlib and from pandas as well
df['Pclass'].value_counts().plot(kind = 'pie',autopct = '%.2f')
```

# 2.Numerical Data

### a.Histogram

```python
plt.hist(df['Age'] )
plt.show()
```

###  b.distplot

```python
sns.distplot(x=df['Age'])
```

### c.Boxplot
Use: Show distribution, outliers, median, and quartiles.
Data Type: Categorical vs Numerical

```python
sns.boxplot(x = df['Age'])
```

```python
df['Age'].skew()#to get the skewness of the data
```

```python
sns.countplot(x = df['Pclass'] , data=df)
```

### ViolinPLot
Use: Combine boxplot and KDE to show distribution.
Data Type: Categorical vs Numerical

```python
sns.violinplot(x =df['Age'] , data=df)
```

```python
df.columns
```

```python
sns.boxplot(x='Fare' , data = df)
```

### Scatterplot
Use: Relationship between two numerical variables.
Data Type: Numerical vs Numerical

```python
df.corr(numeric_only=True)
```

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', style='Sex', alpha=0.7)
```
